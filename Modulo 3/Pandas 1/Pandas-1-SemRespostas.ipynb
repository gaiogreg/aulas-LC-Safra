#### Series 

```python
#Pandas Series
#Um grupo de series forma um dataframe
ds1 = pd.Series(np.random.randint(10, 25, size=25))
ds2 = pd.Series(np.random.randint(10, 25, size=25))
print(type(ds1))
print(ds1.shape)
print(ds1)
print('---------')
print(type(ds2))
print(ds2.shape)
print(ds2)
```

```python
#Matematica
ds = ds1 + ds2
print('Soma:')
print(ds)
print('\n')
print('Subtração:')
ds = ds1 - ds2
print(ds)
print('\n')
print('Multiplicação:')
ds = ds1 * ds2
print(ds)
print('\n')
print("Divisão:")
ds = ds1 / ds2
print(ds)
print('\n')
print('Media:')
print(ds1.mean())
print('\n')
print('Mediana:')
print(ds1.median())
print('\n')
print('Std Dev:')
print(ds1.std())
print('\n')
print('\n')
```

```python
#Comparando series
print("Igual:")
print(ds1 == ds2)
print("Maior que:")
print(ds1 > ds2)
print("Menor que:")
print(ds1 < ds2)
```

##### DYI Series

- Crie duas series aleatorias com 30 numeros, no intervalo 20 a 50 cada  ds1 ds2
- Crie uma nova serie com o nome g_than com a comparação entre a s1 e a s2 
- Crie uma nova serie com o nome ds_sum com a soma das duas series 
- Apresente o desvio padrão da serie ds_sum
- Crie uma nova serie com 30 strings de 5 caracteres aleatorios (Dica. random.choice + string.ascii_lowercase)
- Crie uma lista com todas as series criadas 
- Crie um dataframe com as suas series (Dica. pd.concat())
- Coloque o nome de cada serie no novo dataframe (Dica .rename)

```python
import random 
import string
```

```python
data_set1 = pd.Series(np.random.randint(20, 50, size=30))
data_set2 = pd.Series(np.random.randint(20, 50, size=30))

g_than = data_set1 > data_set2
ds_sum = data_set1 + data_set2
# print(ds_sum.std)

list_str = []
element = ''
for i in range(30):
    for j in range(5):
        element+= random.choice(string.ascii_lowercase)
    list_str.append(element)
    element = ''
ds_str = pd.Series(list_str)

lista_completa = [data_set1, data_set2, g_than, ds_sum, ds_str]

# df_completo = pd.DataFrame(lista_completa)
# df_completo.rename(index={0:'data_set1', 1:'data_set2', 2:'g_than', 3:'ds_sum', 4:'ds_str'}, inplace=True)

df_completo = pd.concat(lista_completa, axis= 1)
df_completo.rename(columns={0:'data_set1', 1:'data_set2', 2:'g_than', 3:'ds_sum', 4:'ds_str'}, inplace=True)
df_completo
```

>Uma função lambda do Python se comporta como uma função normal em relação aos argumentos. Portanto, um parâmetro lambda pode ser inicializado com um valor padrão: o parâmetro **n** toma o **n** externo como valor padrão.
